```python
import pyarrow.parquet as pq
import os

# Diretório que contém os arquivos Delta
directory = "/home/jamilsonfs/airflow/dags/SIA/tmp/DELTA/ACFMG2001.delta/"

# Filtra apenas os arquivos .parquet no diretório
parquet_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.parquet')]

# Lê todos os arquivos Parquet no diretório filtrado
dataset = pq.ParquetDataset(parquet_files)
table = dataset.read()


df = table.to_pandas()
# Exibe a tabela completa
display(df)
```
