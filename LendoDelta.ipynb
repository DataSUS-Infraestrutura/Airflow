# **Implementação**

```python
def drop_anos_diferentes(df):
    df['ano_inter'] = df['DT_INTER'].astype(str).str[:4]
    
    df_filtrado = df[df['ano_inter'].isin(['2020', '2021'])]
    
    return df_filtrado


df_filtrado = drop_anos_diferentes(df)
df_filtrado 
```

```python
df_filtrado['ano_inter'] = df_filtrado['DT_INTER'].astype(str).str[:4]
df_filtrado['ano_saida'] = df_filtrado['DT_SAIDA'].astype(str).str[:4]

anos_comb_filtrados = pd.concat([df_filtrado['ano_inter'], df_filtrado['ano_saida']])

contagem_anos_filtrados = anos_comb_filtrados.value_counts()

display(contagem_anos_filtrados)
```
